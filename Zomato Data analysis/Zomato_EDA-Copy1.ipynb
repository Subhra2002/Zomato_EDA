```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how="left")
```

```python
final_df.Country.value_counts()
```

```python
country_name = final_df.Country.value_counts().index
```

```python
country_name
```

```python
country_val = final_df.Country.value_counts().values
```

```python
country_val
```

```python
plt.pie(country_val[:3],labels=country_name[:3],autopct="%1.2f%%")
```

```python
#Observation: Zomato maximum records or transition are from India after that USA and then United Kingdoms
```

```python
final_df.columns
```

```python
rating = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
rating
```

```python
rating.head()
```

```python
plt.rcParams['figure.figsize']= (18,8)
sns.barplot(x='Aggregate rating',y = 'Rating Count',data = rating)
```

```python
sns.barplot(x='Aggregate rating',y = 'Rating Count',hue = "Rating color",data = rating,palette = ['black','red','orange',"yellow","green","blue"])
```

```python
# From the above graph we observe that:
# 1- Not rated count is very high (black color)
# 2- Maximum number of rating are between 2.5 to 3.4
```

```python
#count plot
sns.countplot(x="Rating color",data = rating,palette = ['black','red','orange',"yellow","green","blue"])
```

```python
### find the countries name that has given zero rating
final_df[final_df['Aggregate rating'] == 0].groupby('Country').size().reset_index()
```
